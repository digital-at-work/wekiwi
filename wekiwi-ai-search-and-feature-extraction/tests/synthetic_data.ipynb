```python
from pydantic import BaseModel, Field
from typing import List
import dspy

# Synthetic data generation using DSPy
lm = dspy.Google(model='gemini-1.5-pro-latest', api_key="AIzaSyAwd02zs-fbHQW5IVyf3vvtoAh31NqDsSA")

dspy.settings.configure(lm=lm, max_tokens=768, init_temperature=0.82)

# Step 1: Define Input and Output Models
class TopicInput(BaseModel):
    topic: str = Field(description="The topic for which to generate the query and texts")

class TopicOutput(BaseModel):
    query: str = Field(description="The generated QUERY, KEYWORD or KEYWORDS")
    pos: List[str] = Field(description="List of generated texts DIRECTLY RELATED to the query")
    neg: List[str] = Field(description="List of generated hard negative texts (RELATED BUT NOT RELEVANT to the query)")

# Step 2: Create Typed Signature
class TopicSignature(dspy.Signature):
    """Generate very high quality GERMAN training data for a BERT model based on the provided topic."""
    input: TopicInput = dspy.InputField()
    output: TopicOutput = dspy.OutputField()

# Step 4: Create an Instance of the Predictor
predictor = dspy.TypedPredictor(TopicSignature)
test = TopicInput(topic="newspaper article on city administration")

prediction = predictor(input=test)

# Step 5: Test the Predictor
#prediction = [TopicInput(topic=topic) for topic in topics]

# Print the output
print(f"Prediction: {prediction}")
print(f"Query: {prediction.query}")
print(f"Positive Texts: {prediction.pos}")
print(f"Negative Texts: {prediction.neg}")
```
